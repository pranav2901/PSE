# MATPLOTLIB tutorial

```python
import matplotlib as pylab
print(pylab.__version__)
```

```python
#Let's create some data for us to use
import numpy as np
x = np.linspace(0, 10, 25) #Data generation for ploting

'''np.linspace() Return evenly spaced numbers over a specified interval.

Returns num evenly spaced samples, calculated over the interval [start, stop].

The endpoint of the interval can optionally be excluded.'''

y = x * x + 2

print(x)
print(y)
```

```python
print(np.array([x,y]))
```

```python
#Anways, let's continue with plotting
#It only take 1 command to plot

import matplotlib.pyplot as plt

plt.plot(x,y, 'r') #here r stands for red color  can try with g b y etc but make sure to use colors specific to
                   #trends or conventions like green for profits and red for losses etc
```

```python
plt.plot(x,y,'r--')
```

```python
#We need separate subplot for every plot on the main canvas else the plots will mess up like in this example
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x, 'g*-')
```

```python
#Now just play around with the numbers to understand how it works
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
```

```python
#So above we see (1,2,1) & (1,2,2) are significant if we want to tile two subplots, otherwise they are plotting in the same
#plot, if we have the same index then they will in one plot only (try it again)

#you can see both of them are having values 0 to 10 on either axis (x or y)
```

```python
fig = plt.figure() #Creating a canvas

axis = fig.add_axes([0.8,0.8,.55,0.9]) #controls the left,bottom,width & height of the canvas between (0 & 1)

axis.plot(x,y,'r')
```

```python
#Now over here also we can draw a subplot, matplotlib.pyplot gives advance techniques to plot than what we have seen earlier
#Let's try this

fig, axes = plt.subplots(nrows = 1, ncols = 3) #Subplot is of 1 row and 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
#We can also picture or graph inside another graph
fig = plt.figure() #creating canvas

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #controls the left,bottom,width & height of the canvas between (0 & 1)
axes2 = fig.add_axes([0.28,0.10,0.20,0.1]) #you have to hit and try for better representation of graphs, no thumb rule

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
#Let's try with different figure size by putting different values to see how canvas size changes

fig = plt.figure(figsize = (6,6)) #Figure size is width and height in inches

axis = fig.add_axes([0.5,0.1,0.8,0.8]) #controls the left,bottom,width & height of the canvas between (0 & 1)

axis.plot(x,y,'r')
```

```python
#Let's try with different dpi size by putting different values to see how long it takes, more dpi value higher memory consumption

fig = plt.figure(figsize = (5,3), dpi = 100) #dpi is dots per inch and makes the picture darker as well

axis = fig.add_axes([0.5,0.1,0.8,0.8]) #controls the left,bottom,width & height of the canvas between (0 & 1)

axis.plot(x,y,'r')
```

```python
#Lets add a title and labels to default plot

fig, axes = plt.subplots()
axes.set_title('Students Details')
axes.set_xlabel('Age')
axes.set_ylabel('Height')
axes.plot(x,y,'r')

#Try giving different labels and title names just to the effect
```

```python
#Lastly let's try to add Legends

#ax.legend(['label1', 'label2']) #just a Syntax

fig, axes = plt.subplots()
axes.set_xlabel('x - Label')
axes.set_ylabel('y - Label')
axes.set_title('title')


axes.plot(x,x**2,'r')
axes.plot(x,x**3,'g')

axes.legend(["y = x**", "y = x***"], loc = 2) #Try changing values(1 to 10) in location to see the effect on legend box position
```

```python
#In matplolib you can set other proprties as well like line color, transparancy and many more, let's see few
fig, axes = plt.subplots(dpi =150)
axes.plot(x,x+1)
axes.plot(x,x+2)
axes.plot(x,x+3)
```

```python
#Setting up color manually with another method
fig, axes = plt.subplots(dpi =150)
axes.plot(x,x+1, color = 'red')
axes.plot(x,x+2, 'blue')
axes.plot(x,x+3, 'green')
```

```python
#Setting up alpha manually for transparancy
fig, axes = plt.subplots(dpi =150)
axes.plot(x,x+1, color = 'red', alpha = 0.5)
axes.plot(x,x+2, 'blue')
axes.plot(x,x+3, 'green', alpha = 0.2)
```

```python
#See the transperancy effect with another way
fig, axes = plt.subplots(dpi =150)
axes.plot(x,x**2, color = 'red', alpha = 1)
axes.plot(x,x/2, 'blue')
axes.plot(x,x+3, 'green', alpha = 0.3)
```

```python
#Linewidth
fig, ax = plt.subplots(dpi=100)

ax.plot(x, x+1, color ='blue', linewidth = .2)
ax.plot(x, x+2, color ='blue', linewidth = .5)
ax.plot(x, x+3, color ='blue', linewidth = 1.5)
ax.plot(x, x+4, color ='blue', linewidth = 5)
```

```python
#Linestyle
fig, ax = plt.subplots(dpi=100)

ax.plot(x, x+1, color ='blue', lw = 2, linestyle = '-')
ax.plot(x, x+2, color ='blue', lw = 2, linestyle = '-.')
ax.plot(x, x+3, color ='blue', lw = 2, linestyle = ':')
ax.plot(x, x+4, color ='blue', lw = 2, ls = '--')
ax.plot(x, x+5, 'g*', lw = 2)
```

```python
#Set Dashes
fig, ax = plt.subplots(dpi=100)

ax.plot(x, x+1, color ='blue', lw = 2, linestyle = '-')
ax.plot(x, x+2, color ='blue', lw = 2, linestyle = '-.')
ax.plot(x, x+3, color ='blue', lw = 2, linestyle = ':')

line, = ax.plot(x, x+4, color = 'black', lw = 1.50)
line.set_dashes([50,20,50,20])
```

```python
#Set the canvas grid and axes limits/range
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize = (10,5)) #remember 2 is an array of axes

axes[0].plot(x, x**2, x, x**3, lw = 2)
axes[0].grid(True)

axes[1].plot(x, x**2, x, x**3, lw = 2)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5])
axes[1].grid(True)
```

# Other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize = (16,5))

axes[0].set_title('Scatter')
axes[0].scatter(x, x + x *0.25*np.random.randn(len(x)))

axes[1].set_title('Step')
axes[1].step(n, n**2, lw =2)

axes[2].set_title('Bar')
axes[2].bar(n, n**2, align = 'center', width = 0.5, alpha = 1) #Try changing width and alpha numbers, also removing all 
                                                                 #additional arameters and see with default settings

axes[3].set_title('Fill_between')
axes[3].fill_between(x, x**2, x**3, color = 'green', alpha = 0.5) #Fill between the two lines
```

```python
#Draw Histogram
#n = np.random.randn(10)
n = [10,12,13,13,14,15,16,16,17,18,19,19,20]
print(n)
print('\n')
fig, axes = plt.subplots(1,2, figsize = (12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n, bins = 5)

axes[1].set_title("Cumulative Detailed Histogram")
axes[1].hist(n, cumulative = True, bins = 5) #Try changing different number for bins also try removing cumulative

#Observe the values on y-axis
```

# Write a python program to plot a pie chart showing popularity of programming languages

```python
import matplotlib.pyplot as plt

# Data to plot

languages = 'Python' , 'Java', 'PHP', 'C#', 'C++'
popularity = [32.5, 10.5, 17, 20, 20]
colors = ['Blue', 'Green', 'Red', 'Yellow', 'Purple']
```

```python
# Explode 1st slice

explode = (0.5, 0, 0, 0, 0)
plt.pie(popularity)
```

```python
plt.pie(popularity, explode = explode, labels = languages, colors = colors, autopct = '%1.1f%%', shadow = True, startangle =0)
```
